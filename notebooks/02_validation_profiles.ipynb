# Schema Validation and Profiles

This notebook demonstrates profile-aware validation using:
- `base`
- `batterypass`


```python
from pathlib import Path
import json
import re
import sys
import subprocess

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    # If notebook is opened from inside notebooks/, move to repo root.
    ROOT = ROOT.parent

assert (ROOT / 'src').exists(), f'Repo root not found from {Path.cwd()}'
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

print('ROOT:', ROOT)

from battinfo.validate.pydantic import validate_json
```

```python
base_doc = json.loads((ROOT / 'src' / 'battinfo' / 'data' / 'examples' / 'cells' / 'A123_20AH.curated.json').read_text(encoding='utf-8'))
base_result = validate_json(base_doc, profile='base')
print('base ok:', base_result.ok)
print('base errors:', base_result.errors[:3])
```

```python
bp_doc = json.loads((ROOT / 'assets' / 'examples' / 'profiles' / 'batterypass-minimal.json').read_text(encoding='utf-8'))
bp_result = validate_json(bp_doc, profile='batterypass')
print('batterypass ok:', bp_result.ok)
print('batterypass errors:', bp_result.errors[:3])
```

```python
broken = json.loads((ROOT / 'assets' / 'examples' / 'profiles' / 'batterypass-minimal.json').read_text(encoding='utf-8'))
broken.pop('measurements', None)
res = validate_json(broken, profile='batterypass')
print('broken ok:', res.ok)
print('first error:', res.errors[0] if res.errors else None)
```

```python
res = validate_json(base_doc, profile='unknown-profile')
print('ok:', res.ok)
print('error:', res.errors[0])
```

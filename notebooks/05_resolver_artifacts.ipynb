# Resolver Artifact Build and Inspection

This notebook demonstrates how BattINFO resources are prepared for `w3id` resolution.

It covers:
- building `registry/site` artifacts
- inspecting generated HTML/JSON/JSON-LD
- checking the `w3id` rewrite template


```python
from pathlib import Path
import json
import re
import sys
import subprocess

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    # If notebook is opened from inside notebooks/, move to repo root.
    ROOT = ROOT.parent

assert (ROOT / 'src').exists(), f'Repo root not found from {Path.cwd()}'
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

print('ROOT:', ROOT)
```

```python
result = subprocess.run(
    [sys.executable, 'scripts/build_resolver_artifacts.py'],
    cwd=ROOT,
    capture_output=True,
    text=True,
)
print(result.stdout.strip())
if result.stderr:
    print(result.stderr)
print('exit code:', result.returncode)
```

```python
site = ROOT / 'registry' / 'site'
for path in sorted(site.rglob('*')):
    if path.is_dir():
        print(path.relative_to(ROOT))
```

```python
sample_jsonld = ROOT / 'registry' / 'site' / 'cell' / '3m6k-9t2p-7x4h-9nq8' / 'index.jsonld'
sample_html = ROOT / 'registry' / 'site' / 'cell' / '3m6k-9t2p-7x4h-9nq8' / 'index.html'
print('jsonld exists:', sample_jsonld.exists())
print('html exists:', sample_html.exists())
if sample_jsonld.exists():
    obj = json.loads(sample_jsonld.read_text(encoding='utf-8'))
    print('jsonld @id:', obj.get('@id'))
    print('jsonld @type:', obj.get('@type'))
```

```python
w3id_template = (ROOT / 'registry' / 'w3id-template.htaccess').read_text(encoding='utf-8')
print(w3id_template[:1200])
```
